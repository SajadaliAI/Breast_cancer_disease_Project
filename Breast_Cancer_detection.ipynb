```python
# step 7 >>> split and train model!
```

```python
x = cancer_df.drop("target",axis =1)
y = cancer_df["target"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
# step8>> Scale features
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_scale = scale.fit_transform(x_train)
y_scale = scale.fit_transform(x_test)
```

```python
# step9>>  model_selection using Different Algorithms
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
naive_model     =GaussianNB()
naive_model.fit(x_train,y_train)
y_predicted_nv = naive_model.predict(x_test)
print("Accuracy_score of naive_model is: ",accuracy_score(y_test,y_predicted_nv))
```

```python
tree_model      =DecisionTreeClassifier(criterion='gini',max_features=0.6)
tree_model.fit(x_train,y_train)
y_predicted_tree = tree_model.predict(x_test)
print("Accuracy_score tree_model is: ",accuracy_score(y_test,y_predicted_tree))
```

```python
logist_model    =LogisticRegression(max_iter=100,penalty='l2')
logist_model.fit(x_train,y_train)
y_predicted_logi = logist_model.predict(x_test)
print("Accuracy_score logist_model is: ",accuracy_score(y_test,y_predicted_logi))
```

```python
random_model    =RandomForestClassifier(criterion='gini',n_estimators=10)
random_model.fit(x_train,y_train)
y_predicted_rand = random_model.predict(x_test)
print("Accuracy_score random_model is: ",accuracy_score(y_test,y_predicted_rand))
```

```python
ada_boost_model =AdaBoostClassifier(n_estimators=10,
    algorithm='SAMME.R',
    random_state=10,
)
ada_boost_model.fit(x_train,y_train)
y_predicted_ada = ada_boost_model.predict(x_test)
print("Accuracy_score ada_boost_model is: ",accuracy_score(y_test,y_predicted_ada))
```

```python
xgb_model       =XGBClassifier()
xgb_model.fit(x_train,y_train)
y_predicted_xgb = xgb_model.predict(x_test)
print("Accuracy_score xgb_model is: ",accuracy_score(y_test,y_predicted_xgb))
```

```python
# step10>> Confusion Matrix
```

```python
cm = confusion_matrix(y_test,y_predicted_xgb)
sns.heatmap(cm,annot=True)
plt.show()
```

```python
# step11>> Classification_report
```

```python
print(classification_report(y_test,y_predicted_xgb))
plt.show()
```

```python
# step12>> Save Model!!
```

```python
import pickle 

pickle.dump(xgb_model,open("breast_cancer_detecter.pkl","wb"))
breast_cancer_model = pickle.load(open("breast_cancer_detecter.pkl","rb"))

y_pred = breast_cancer_model.predict(x_test)

print("Accuracy_score of XGB_model is: ",accuracy_score(y_test,y_pred))
print("Confusion_matrix of XGB_model is: ","\n",confusion_matrix(y_test,y_pred),"\n")
```
